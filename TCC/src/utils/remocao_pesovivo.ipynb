```python
# Tratar os floats que foram inseridos com virgula ao inves de ponto.

def replaceFloat(nameColumns):
    aux = [float(str(x).replace(',', '.')) for x in df[nameColumns]]
    return aux


# for x in df:
#     df[x] = replaceFloat(x)

df['PESOVIVO'] = replaceFloat('PESOVIVO')

df
```
